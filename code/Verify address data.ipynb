```python
import pandas as pd
import requests
```

```python
df = pd.read_csv("https://github.com/uva-bi-sdad/national_address_database/raw/main/data/VA_Arlington.csv.xz")
df
```

```python
pdf = df.sample(3)
pdf
```

```python
pdf['response'] = pdf['address'].apply(lambda x: requests.get('https://geocoding.geo.census.gov/geocoder/locations/onelineaddress?address=%s&benchmark=2020&format=json' % x).json())
```

```python
pdf['response']
```

```python
pdf['valid'] = pdf['response'].apply(lambda x: len(x['result']['addressMatches']) > 0)
```
